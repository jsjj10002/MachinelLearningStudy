### 데이터셋 분할 

```python
#데이터 생성
import numpy as np
from sklearn.model_selection import train_test_split

X,y = np.arange(120).reshape((30,4)), list(range(30))
print('X의 첫 5개의 샘플:\n', X[:5,:],'\n')
print('y의 첫 5개의 샘플:\n', y[:5])
```

```python
#X와 y를 각각 학습 데이터와 테스트 데이터로 분할, 테스트 비율; 33%, random_state에 값 설정해 랜덤성 발생 하지 않음
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state=1234)
print('데이터 분할:', len(X_train),len(y_train), len(X_test), len(y_test),'\n')
print('X_trani의 첫 5개의 샘플:\n', X_train[:5,:],'\n')
print('y_train의 첫 5개의 샘플:\n', y_train[:5])
```

```python
#검증데이터 정의- 학습데이터를 다시 학습데이터와 검증데이터로 나눔 
X_train, X_val, y_train, y_val = train_test_split(X_train,y_train,test_size=0.33, random_state=1234)
print('데이터 분할:', len(X_train),len(y_train), len(X_val), len(y_val), len(X_test),len(y_test),'\n')
print('X_trani의 첫 5개의 샘플:\n', X_train[:5,:],'\n')
print('y_train의 첫 5개의 샘플:\n', y_train[:5])
```
